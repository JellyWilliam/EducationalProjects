### 5. Посчитайте для каждого пользователя:

#### Количество сделанных звонков и израсходованных минут разговора по месяцам;

```python
calls_min_1_minutes = calls.copy(deep=True)
calls_min_1_minutes.loc[calls_min_1_minutes["duration"] < 1] = 1
calls_min_1_minutes["call_date"] = pd.to_datetime(calls_min_1_minutes["call_date"], errors='coerce').dt.to_period('m')
calls_count = calls_min_1_minutes.groupby(["user_id", "call_date"])["id"].count().reset_index().set_index(["user_id", "call_date"])
calls_spent_minutes = calls_min_1_minutes.groupby(["user_id", "call_date"])["duration"].sum().reset_index().set_index(["user_id", "call_date"])
calls_count.columns = ["calls_count"]
calls_spent_minutes.columns = ["calls_duration"]
```

```python
calls_count_spent_minutes = pd.concat([calls_count, calls_spent_minutes], axis=1, join="inner")
calls_count_spent_minutes
```

#### Количество отправленных сообщений по месяцам;

```python
messages_every_mounth = messages.copy(deep=True)
messages_every_mounth["message_date"] = pd.to_datetime(messages_every_mounth["message_date"], errors='coerce').dt.to_period('m')
count_messages_every_mounth = messages_every_mounth.groupby(["user_id", "message_date"])["id"].count().reset_index().set_index(["user_id", "message_date"])
count_messages_every_mounth.columns = ["messages_count"]
count_messages_every_mounth
```

#### Объем израсходованного интернет-трафика по месяцам;

```python
internet_every_mounth = internet.copy(deep=True)
internet_every_mounth["session_date"] = pd.to_datetime(internet_every_mounth["session_date"], errors='coerce').dt.to_period('m')
internet_used_every_mounth = internet_every_mounth.groupby(["user_id", "session_date"])["mb_used"].sum().reset_index().set_index(["user_id", "session_date"])
internet_used_every_mounth.columns = ["internet_used"]
internet_used_every_mounth
```

#### Сводная таблица

```python
user_every_mounth_all_used = pd.concat([calls_count_spent_minutes, count_messages_every_mounth, internet_used_every_mounth], axis=1, join="inner")
user_every_mounth_all_used
```

#### Помесячную выручку с каждого пользователя (вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика;
#### Остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану).

```python
for user in user_every_mounth_all_used.index:
    user_every_mounth_all_used.loc[user, 'tariff'] = users.loc[user[0]-1000, 'tariff']
user_every_mounth_all_used
```

```python
def det_revenue(row):
    messages = row['messages_count']
    mb_used = row['internet_used']
    tariff = row['tariff']
    duration = row['calls_duration']
    calls = row['calls_count']
        
    extra_duration = duration - tariffs.loc[tariff, 'minutes_included']
    extra_mb = mb_used - tariffs.loc[tariff, 'mb_per_month_included']
    extra_messages = messages - tariffs.loc[tariff, 'messages_included']

    if extra_duration < 0: extra_duration = 0
    if extra_mb < 0: extra_mb = 0
    if extra_messages < 0: extra_messages = 0

    return tariffs.loc[tariff, 'rub_per_message'] * extra_messages + (tariffs.loc[tariff, 'rub_per_gb'] / 1024) * extra_mb + tariffs.loc[tariff, 'rub_per_minute'] * extra_duration + tariffs.loc[tariff, 'rub_monthly_fee']
```

```python
user_every_mounth_all_used['revenue'] = user_every_mounth_all_used.apply(det_revenue, axis=1)
user_every_mounth_all_used
```

## Шаг 4. Проверьте гипотезы

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;

$H_{0}$  — средняя выручка пользователей тарифа «Ультра» равна средней выручке пользователей тарифа «Смарт».

```python
smart = user_every_mounth_all_used.query('tariff == 0')['revenue']
ultra = user_every_mounth_all_used.query('tariff == 1')['revenue']
alpha = .05
```

```python
user_every_mounth_all_used.pivot_table(index='tariff', values='revenue', aggfunc='median')
```

```python
results = stats.ttest_ind(
    ultra, 
    smart, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

$H_{0}$  — средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов.

```python
for user in user_every_mounth_all_used.index:
    user_every_mounth_all_used.loc[user, 'city'] = users.loc[user[0]-1000, 'city']

moscow = user_every_mounth_all_used.query('city == "Москва"')['revenue']
regions = user_every_mounth_all_used.query('city != "Москва"')['revenue']
```

```python
def det_region(city):
    if city == 'Москва': return 'Москва'
    return 'Другой регион'
    
user_every_mounth_all_used['region'] = user_every_mounth_all_used['city'].apply(det_region)
user_every_mounth_all_used.pivot_table(index='region', values='revenue', aggfunc='median')
```

```python
results = stats.ttest_ind(
    moscow, 
    regions, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
